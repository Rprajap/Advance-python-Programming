Q.1. For this challenge, forget how to add two numbers together. The best explanation on what to do for this function is this meme: 


Examples

meme_sum(26, 39) ➞ 515\

2+3 = 5, 6+9 = 15\
26 + 39 = 515
meme_sum(122, 81) ➞ 1103\

1+0 = 1, 2+8 = 10, 2+1 = 3\
122 + 81 = 1103
meme_sum(1222, 30277) ➞ 31499

Answer :-

```python
def meme_sum(a,b):
    a = str(a)
    b = str(b)
    sum1 = ''
    
    while True:
        if len(a) < len(b):
            a = '0'+a
        elif len(b) < len(a):
            b = '0'+b
        else:
            break
    
    for i in range(len(a)):
        sum1 += str(int(a[i]) + int(b[i]))

    return sum1

print(meme_sum(26, 39))
print(meme_sum(122,81))
print(meme_sum(1222, 30277))
print(meme_sum(10,203))
```

**Q2**. Given an integer, create a function that returns the next prime. If the number is prime, return the number itself.

**Examples**:-

next_prime(12) ➞ 13

next_prime(24) ➞ 29

next_prime(11) ➞ 11\

11 is a prime, so we return the number itself.

**Answer**:-

```python
def prime_num(n):
    for i in range(2,n):
        if n % i == 0:
            break
    else:
        return n
        
def next_prime(n):
    while True:
        if prime_num(n) != None:
            break
        n = n + 1
        prime_num(n)
    
    return prime_num(n)

print(next_prime(12))
print(next_prime(24))
print(next_prime(11))
print(next_prime(83))
```

**Q3**. If a person traveled up a hill for 18mins at 20mph and then traveled back down the same path at 60mph then their average speed traveled was 30mph. Write a function that returns the average speed traveled given an uphill time, uphill rate and a downhill rate. Uphill time is given in minutes. Return the rate as an integer (mph). No rounding is necessary.

**Examples**:-

ave_spd(18, 20, 60) ➞ 30

ave_spd(30, 10, 30) ➞ 15

ave_spd(30, 8, 24) ➞ 12

**Answer**:-

```python
# as distance is same, formula  is  2ab/a+b where a and b are two speeds

def ave_spd(uphill_time, uphill_rate, downhill_rate):
    return int(2 * uphill_rate * downhill_rate/ (uphill_rate + downhill_rate))

print(ave_spd(18, 20, 60))
print(ave_spd(30, 10, 30))
print(ave_spd(30,8,24))
```

**Q4**. The Kempner Function, applied to a composite number, permits to find th smallest integer greater than zero whose factorial is exactly divided by the number.

kempner(6) ➞ 3 1! = 1 % 6 > 0 2! = 2 % 6 > 0 3! = 6 % 6 === 0

kempner(10) ➞ 5 1! = 1 % 10 > 0 2! = 2 % 10 > 0 3! = 6 % 10 > 0 4! = 24 % 10 > 0 5! = 120 % 10 === 0

A Kempner Function applied to a prime will always return the prime itself.

kempner(2) ➞ 2

kempner(5) ➞ 5

Given an integer n, implement a Kempner Function.

**Examples**:-

kempner(6) ➞ 3

kempner(10) ➞ 5

kempner(2) ➞ 2

**Answer**:-

```python
#n = 6

def fact(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * fact(n-1)
        
def kempner(n):
    for i in range(1,n+1):
        if fact(i) % n == 0:
            return i

print(kempner(6))
print(kempner(10))
print(kempner(2))
```
